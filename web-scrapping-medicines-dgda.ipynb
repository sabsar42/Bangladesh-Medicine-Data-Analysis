# Allopathic using Selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import json
import time

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')

driver = webdriver.Chrome(options=options)

url = "https://dgdagov.info/index.php/registered-products/allopathic"
driver.get(url)

wait = WebDriverWait(driver, 20)  # Increase the timeout to 20 seconds

def extract_data_from_page():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table', {'id': 'gridData'})
    table_header = soup.find('table', {'class': 'display dataTable'})

    if table:
        header_rows = table_header.find_all('tr')
        headers = [header.text.strip() for header in header_rows[0].find_all('th')]

        rows = table.find_all('tr')
        extracted_data = []
        for row in rows[1:]:
            columns = row.find_all('td')
            row_data = {headers[i]: column.text.strip() for i, column in enumerate(columns)}
            extracted_data.append(row_data)
            print(extracted_data)
        return extracted_data
    else:
        return []

all_data = []

# Scroll down on the first page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)  # Adjust the sleep time as needed

while True:
    data_from_page = extract_data_from_page()
    if not data_from_page:
        break 

    all_data.extend(data_from_page)

    print(f"Data from current page: {data_from_page}")

    next_page_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//a[text()="Next"]')))

    if not next_page_button:
        break

    next_page_button.click()
    # Scroll down to trigger loading of additional content
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait for the new content to load
    time.sleep(5)  # Adjust the sleep time as needed

driver.quit()
result = {'allopathic': all_data}

json_file_path = '/kaggle/working/allopathic.json'
with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(result, json_file, indent=2)

print(f"Data has been saved to '{json_file_path}'")
```
